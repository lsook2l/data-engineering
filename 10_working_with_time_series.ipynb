### <strong> Python 내장 모듈을 이용한 Datetime 생성
---

#### ```datetime```, ```dateutil```: 시계열 처리를 위한 Python 내장 모듈

````python
from datetime import datetime

# datetime 객체 생성
t = datetime(year=2025, month=5, day=7) # 매개변수 삽입
t # 시,분은 입력 안 함 -> 기본값으로 출력됨
````

````python
type(t)
````

````python
# datetime 속성 접근: year
t.year # 객체의 속성들 가져올 수 있음 #t.month 등
````

````python
# 문자열 파싱을 통한 날짜 해석
from dateutil import parser # parser.parse : 다양한 방식의 날짜 데이터를 '해석'해서 공통 데이터 타입으로 맞춰줌
# 2024-11-05, 11-05, 2024 # 영어권에사 날짜 표현 방식
date = parser.parse("7th of May, 2025") # 문자열 
# date = parser.parse("05-07-2025")

date
````

#### ```strftime()```: datetime의 속성을 문자열 형태로 출력

````python
# [+] 요일 출력
date.strftime('%A') # %a = 'wed'
````

````python
# [+] 월 출력
date.strftime('%B')
````

### <strong> NumPy를 이용한 Datetime 배열 생성</strong>
---
```datetime64```: ```NumPy```에서 지원하는 datetime 클래스

````python
import numpy as np
# dates = np.array(['2025-05-07', '2025-05-08']) -> type = 문자열

# datetime64 배열 객체 생성
dates = np.array(['2025-05-07', '2025-05-08'], dtype=np.datetime64) # datetime64 객체로 시간 데이터 표현
dates
````

````python
# datetime64() 메서드를 이용한 객체 생성
date = np.datetime64('2025-05-01')
date
````

````python
np.arange(31) # 백터
````

````python
# 벡터화 연산을 통한 Datetime 배열 생성
date + np.arange(31) # [0, 1, .. 30] # 스칼라 + 백터 = 브로드캐스팅이 이루어짐 # '일'로 표현됨
````

````python
# 분(minute) 단위의 Datetime 객체 생성
t = np.datetime64('2025-05-07 14:30')
t
````

````python
# 디폴트 시간 빈도: ms까지 나옴
t = np.datetime64('2025-05-07 14:30:59.99')
t
````

````python
# 시간 정밀도를 Nanosecond로 설정
t = np.datetime64('2025-05-07 14:30:59.99', 'ns')
t
````

````python
# datetime64 코드: Y, M, D, h...
t = np.datetime64('2025-05-07 14:30:59.99', 'M') # 월 단위로 변경하겠다
t
````

### <strong> Pandas에서의 시계열 처리</strong>
---

````python
import pandas as pd

# [+] to_datetime() 함수를 이용한 Timestamp 객체 생성 
date = pd.to_datetime("2004-02-03") # 생일로 입력
date
````

````python
type(date)
````

````python
# [+] 요일 출력
date.strftime('%A')
````

````python
# date + np.arange(7) # 객체 유형이 달라서 안 됨
````

````python
# NumPy 스타일의 배열 연산
date + pd.to_timedelta(np.arange(7)) # 넘파일 배열의 타임스탬프로 변경해줌 - 타임스탬프의 배열을 표현하는 객체 유형
````

````python
date + pd.to_timedelta(np.arange(7), unit='D') # datetime 객체
````

#### <strong> 시계열 관련 Pandas의 데이터 구조

````python
# Timestamp와 DatetimeIndex #
dates = pd.to_datetime([datetime(2015, 7, 3),
                      "4th of July, 2015",
                      '2015-Jul-6',
                      '07-07-2015',
                      '20150708']) # 다 다른 포맷을 시간 표현한 상태
print(dates) # 정확한 시점만 표현
````

````python
# to_period(): DatetimeIndex -> PeriodIndex 변환
dates.to_period('D') # 비슷해보이지만 시작과 종료 정보 포함 # 따라서 'W'로 주로 표현
````

````python
# [+] TimedeltaIndex 생성
dates - dates[0] # 시간차 정보 구해줄 수 있음
````

#### <strong> 정규 시퀀스 </strong>

````python
# [+] pd.date_range()를 이용한 정규 시퀀스 생성
pd.date_range('2026-01-01', '2026-04-28') # '일'단위로 고정된 빈도로 타임스탬프 생성
````

````python
# [+] Period 값을 이용
pd.date_range('2026-04-28', periods=100) # 4월 28일부터 100일간의 타임스탬프 생성
````

````python
# 빈도 변경 (시간 데이터의 반복 주기)
pd.date_range('2022-10-27', periods=12, freq='h') 
````

````python
# pd.period_range()
pd.period_range('2015-07', periods=8, freq='M')
````

````python
# Timestamp vs. Period
period = pd.Period('2022-10-27') # 시작 시간-종료 시간
timestamp = pd.Timestamp('2022-10-27 09:30')

print(period.start_time < timestamp < period.end_time) 

print(period.start_time) # 시작 시간

print(period.end_time) # 종료 시간
````

````python
# pd.timedelta_range()
pd.timedelta_range(0, periods=10, freq='h')
````

| Code   | Description         | Code   | Description          |
|--------|---------------------|--------|----------------------|
| ``D``  | Calendar day        | ``B``  | Business day         |
| ``W``  | Weekly              |   -    |                      |
| ``M``  | Month end           | ``BM`` | Business month end   |
| ``Q``  | Quarter end         | ``BQ`` | Business quarter end |
| ``A``  | Year end            | ``BA`` | Business year end    |
| ``H``  | Hours               | ``BH`` | Business hours       |
| ``T``  | Minutes             |   -    |                      |
| ``S``  | Seconds             |   -    |                      |
| ``L``  | Milliseonds         |   -    |                      |
| ``U``  | Microseconds        |   -    |                      |
| ``N``  | nanoseconds         |   -    |                      |

#### <strong> 빈도 및 오프셋

````python
# 빈도 코드
pd.date_range('2022-11-01', periods=10, freq='h')
````

````python
# 빈도, M: 월말
(pd.date_range('2022-11-01', periods=10, freq="ME"))
````

````python
# 오프셋
from pandas.tseries.offsets import DateOffset
ts = pd.Timestamp('2023-11-07 09:30:00')
ts + DateOffset(hours=-40)
````

````python
# 오프셋(minus)
ts - DateOffset(years=73, months=4, days=12, hours=5, minutes=10, seconds=00)
````
